```python
import numpy as np
import pandas as pd
import sklearn
from sklearn.datasets import load_boston
df=load_boston()
print(df.keys())
print(df.target)
boston=pd.DataFrame(df.data,columns=df.feature_names)
(boston.head(10))
boston['Predict']=df.target
(boston.head())
print(boston.isnull())
print(boston.isnull().sum())

x=boston.drop('Predict',axis=1)
y=boston['Predict']

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.14,random_state=5)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
lin_model=LinearRegression()
lin_model.fit(x_train,y_train)

y_train_predict=lin_model.predict(x_train)
rmse1=(np.sqrt(mean_squared_error(y_train,y_train_predict)))
print(">>>>>the model performance for training set")
print('RMSE is {}'.format(rmse1))
print("\n")

y_test_predict=lin_model.predict(x_test)
rmse2=(np.sqrt(mean_squared_error(y_test,y_test_predict)))
print(">>>>>the model performance for testing set")
print('RMSE is {}'.format(rmse2))
print("\n")
print("the difference is",rmse1-rmse2)
```
